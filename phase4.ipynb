```python
x=df[['STATE/UT WISE','TOTAL IPC CRIMES', 'TOTAL SC CRIMES','TOTAL ST CRIMES', 'TOTAL CHILDREN CRIME','Total Property stolen / recovered property Crime','TOTALPOLICEMEN KILLED','Total Place of occurance crime','TOTAL WOMEN CRIMES']].copy()
x
```

```python
from sklearn.cluster import KMeans
```

```python
x
```

```python
x.columns
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
le_st=le.fit_transform(df['STATE/UT WISE'])
le_st
```

```python
df['STATE/UT WISE']=le_st
```

```python
df
```

```python
x['STATE/UT WISE']=le_st
```

```python
x
```

```python
wcss1=[]
```

```python
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,random_state=42)
    kmeans.fit(x)
    wcss1.append(kmeans.inertia_)
plt.plot(range(1,11),wcss1)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('wcss')
plt.show()
```

```python
kmeans=KMeans(n_clusters =3,random_state=42)
y_means=kmeans.fit_predict(x)
print(y_means)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x,y_means) # we have got 66% accuracy.
```
