```python
# Reading the excel file which contains details 
df=pd.read_excel('Phase 1 .xlsx',sheet_name=0)
df
```

```python
df.drop(columns='SNO',axis=1,inplace=True)# Dropping the SNO Column.
```

```python
# To see how much Area ,each state is covering In India,we  use pie Chart.

plt.figure(figsize=(15,35))
plt.pie(df['AREA(in Km2)'],autopct ='% 1.1f %%', shadow = True,labels=df['STATE '])
plt.show()
```

```python
#observations: 
# we have Rajasthan as the First state where Area which occupied (10.8%) in India.
# followed by Madhya pradesh and Maharastra with (9.7 %) Area occupiedin India.
# The One with Least Area occupancy In India is State Goa  with (0.3% )In India
# The One with Least UT ocuupancy in India is Daman & Diu.

```

```python

# To See the State whose Population is High In India:

plt.figure(figsize=(15,20))
plt.pie(df['POPULATION'],autopct ='% 2.3f %%', shadow = True,labels=df['STATE '])
plt.show()
```

```python
# Observations:
# The state with highest population among India is UttarPradesh(15 %) among total population.
#  The state with Lowest Population is Sikkim in India.
```

```python
df.columns
```

```python
# To see the Literacy Rate In India:
plt.figure(figsize=(10,12))
sns.barplot(x=df['LITERACY RATE(in %)'],y=df['STATE '])
plt.show()
```

```python
# observations:
# The state with Highest Literacy rate in India is kerala(0.94%)
# The state with Lowest Literacy rate in India is Bihar (0.62%)
```

```python
# To compare Male and female literates in each state in India.
plt.figure(figsize=(15,20))
sns.barplot(x=df['MALE  LITRATES(in %)'],y=df['FEMALE LITERATES(in %'],hue=df['STATE '])
plt.xlabel('MaleLiteracypercentage')
plt.ylabel('FemaleLiteracypercentage')
plt.title('literacycomparision')
plt.show()
```

```python
# Bihar Female Literacy is less when compared to other states
# kerala Male literacy rate is high compared to the other states.
```

```python
# To See the density of the state in India.
plt.figure(figsize=(20,9))
sns.barplot(x='STATE ',y='DENSITY(km)',data=df,palette='rainbow')
plt.xticks(rotation=30)
```

```python
# The Density of Chandigarh and Delhi is High among the States in India
```
